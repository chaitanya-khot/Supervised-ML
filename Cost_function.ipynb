 # Cost functions
 Cost functions are used to estimate how badly models are performing. Put simply, a cost function is a measure of how wrong the model is in terms of its ability to estimate the relationship between X and y. Alternatively, it tells us how well the model is doing (checks the model's predicted values to the actual values), so that we can try to make it do better. <br>
 (ref: https://towardsdatascience.com/machine-learning-fundamentals-via-linear-regression-41a5d11f5220)

### Formula for cost function: <br>
$f_{w, b}$(x) = wx + b
<br>
Where, <br> 
w, b: parameters / coefficients / weights

For linear regression: we find a value of w such that it minimizes the cost function J(w) <br>
General case: we find a value of w, b such that it minimizes the cost function J(w, b) <br>
the model fit the data relatively well, compared to other choices for parameter w When the cost J is at or near a minimum.
